```python
import pandas as pd 
from tqdm import tqdm
import warnings
warnings.filterwarnings('ignore')

def list_all_files(rootdir, ext=None, excpt=None):
    import os
    _files = []
    list = os.listdir(rootdir) #列出文件夹下所有的目录与文件
    for i in range(0,len(list)):
        path = os.path.join(rootdir,list[i])
        if os.path.isdir(path) and (excpt == None or os.path.basename(path) not in excpt):
            _files.extend(list_all_files(path, ext))
        if os.path.isfile(path) and (ext == None or ext in path):
            _files.append(path)
    return _files

def landingPoint(phaseList,flag):
    id = -1
    if len(flag)==1:
        phase = 'FLARE'
    else:
        phase = 'FINAL APP'
    for i in range(len(phaseList)-1):
        if phaseList[i] == phase and phaseList[i+1] == 'LANDING':
            id = i
            break
    return id

def ldgTransfer(list):
    newList = []
    for i in range(len(list)):
        if list[i] == 'GROUND':
            newList.append(0)
        else:
            newList.append(1)
    return newList

def fillNan(dataList):
    import numpy as np
    newList = []
    index = 0
    while np.isnan(float(dataList[index])): #找到第一个非空元素下标
        index += 1
    for i in range(len(dataList)):
        if np.isnan(float(dataList[i])):
            newList.append(dataList[index])
        else:
            newList.append(dataList[i])
            index = i
    return newList
                                     
def getNameList(name, n):
    nameList = [name]
    for i in range(1, n):
        nameList.append(name + "." + str(i))
    return nameList

def getDataList(df, name, n):
    nameList = getNameList(name, n)
    dataList = []
    data = df[nameList].values
    for i in range(0, len(data)):
        for j in range(n):
            dataList.append(data[i][j])
    return dataList

def getDataList1(df, name, n, flag):
    if len(flag)==1:
        nameList = getNameList(name, n)
    else:
        nameList = getNameList(name, int(n/2))
    data = df[nameList].values
    dataList = []
    for i in range(0, len(data)):
        k = 0
        for j in range(n):
            if len(flag)==1:
                dataList.append(data[i][j])
            else:
                if j%2==0:
                    dataList.append(data[i][k])
                else:
                    dataList.append(data[i][k])
                    k+=1
    return dataList

def toFloat(dataList):
    newList = []
    for data in dataList:
        newList.append(float(data))
    return newList

def majorOperator(list1, list2):
    import numpy as np
    l1 = np.abs(np.array(list1))
    l2 = np.abs(np.array(list2))
    if sum(l1) >= sum(l2):
        return 'First'
    else:
        return 'Second'
    
def readCsv(path):
    import pandas as pd
    import re
    pathFind = path.split('/')
    index = len(pathFind)-1
    pathFind = pathFind[index]
    pattern = 'B-'
    flag = re.findall(pattern,pathFind)
    f = open(path)
    if len(flag)==1:
        df = pd.read_csv(f, skiprows=[0, 1, 2, 3, 4, 6, 7, 8])
    else:
        df = pd.read_csv(f)
        columns = ['FRAME', 'ARN', 'FLIGHT_NO', 'BJ_DATE', 'BJ_TIME', 'CURRENT_DATE', 'DEP_BJ_DATE', 'DEP', 'ARR', 'DEP_BJ_TIME', 'FLIGHT_PHASE', 'ALT_STD', 'ALT_QNH', 'ALT_AFE', 'RADIO_H', 'RALT_LH', 'RALT_LH.1', 'RALT_LH.2', 'RALT_LH.3', 'RALT_RH', 'RALT_RH.1', 'RALT_RH.2', 'RALT_RH.3', 'LDG_STATUS', 'LDG_STATUS.1', 'LDG_STATUS.2', 'LDG_STATUS.3', 'LDGL', 'LDGL.1', 'LDGL.2', 'LDGL.3', 'LDGNOS', 'LDGNOS.1', 'LDGNOS.2', 'LDGNOS.3', 'LDGR', 'LDGR.1', 'LDGR.2', 'LDGR.3', 'IAS', 'VAPP', 'GS', 'VRTG', 'VRTG.1', 'VRTG.2', 'VRTG.3', 'VRTG.4', 'VRTG.5', 'VRTG.6', 'VRTG.7', 'IVV_CA', 'PITCH', 'PITCH.1', 'PITCH.2', 'PITCH.3', 'PITCH_CPT', 'PITCH_CPT.1', 'PITCH_CPT.2', 'PITCH_CPT.3', 'PITCH_FO', 'PITCH_FO.1', 'PITCH_FO.2', 'PITCH_FO.3', 'ROLL', 'ROLL.1', 'ROLL_CPT', 'ROLL_CPT.1', 'ROLL_CPT.2', 'ROLL_CPT.3', 'ROLL_FO', 'ROLL_FO.1', 'ROLL_FO.2', 'ROLL_FO.3', 'APS', 'AP_EGD1', 'AP_EGD2', 'ATS', 'N11', 'N12', 'TLA1', 'TLA2', 'FLAP_SELECT', 'FLAP_POS_LH', 'FLAP_POS_RH', 'FLAPC', 'HEAD_MAG', 'HEAD', 'WIN_DIR', 'WIN_SPD', 'RUDD', 'RUD_PEDAL_POS', 'DME1DSTD', 'DME2DSTD', 'ILS_FRQ1', 'ILS_FRQ2', 'LONPC', 'LATPC', 'GW_C', 'GLIDE_DEV1_C', 'LOC_DEV1_C', 'LATG_C', 'LATG_C.1', 'LATG_C.2', 'LATG_C.3', 'LONG_C', 'LONG_C.1', 'LONG_C.2', 'LONG_C.3', 'FLEX_TEMP1', 'FLEX_TEMP2', 'FLEX_TO_SEL1', 'FLEX_TO_SEL2', 'STAB', 'STAB.1', 'SPOIL_LOUT1', 'SPOIL_ROUT1', 'D_SPD', 'SPD_35FT', 'ELEV_LH', 'ELEV_RH', 'BK_L', 'BK_R', 'ABK_OFF', 'STEER_POS_1', 'STEER_POS_2', 'DIS_THR_TD', 'WHL_SPD_1', 'WHL_SPD_2', 'WHL_SPD_3', 'WHL_SPD_4', 'ATHR_ENG', 'LAT_FMA', 'LON_FMA', 'DME_1_F', 'DME_2_F', 'COG', 'ALT_GPS', 'AOA_L', 'AOA_R', 'EPR_1', 'EPR_2', 'AC_TYPE']
        df.columns = columns
    f.close()
    return df,flag
    
def computeInfo(path):
    import pandas as pd 
    import numpy as np
    
    #pathsplit = path.split('/')
    #file = pathsplit[len(pathsplit)-1]
    dic = {}
    df,flag = readCsv(path)
    phaseList = list(df['FLIGHT_PHASE'])
    point = landingPoint(phaseList,flag) #飞机着陆时间点（精度：秒）
    df = df[point-80 : point+20]
    
    dic['LDGL'] = ldgTransfer(getDataList(df, 'LDGL', 4)) #左起落架状态
    dic['LDGNOS'] = ldgTransfer(getDataList(df, 'LDGNOS', 4)) #右起落架状态
    dic['LDGR'] = ldgTransfer(getDataList(df, 'LDGR', 4)) #前起落架状态
    
    dic['RADIO_LH'] = toFloat(getDataList(df, 'RALT_LH', 4))#左无线电高度
    dic['RADIO_RH'] = toFloat(getDataList(df, 'RALT_RH', 4))#右无线电高度
    ralt_mean = list((np.array(dic['RADIO_LH']) + np.array(dic['RADIO_RH'])) / 2)
    dic['RALT_MEAN'] = ralt_mean
    diffList = list(np.abs(np.array(ralt_mean)-50))
    flare_point = 2 * diffList.index(min(diffList)) #无线电高度50英尺点
    
    alt_std = list(df['ALT_STD'])  # 海拔高度
    alt_qnh = list(df['ALT_QNH'])  # 海拔高度
    off_std = alt_std[int(flare_point/8)] - ralt_mean[int(flare_point/2)]  #海拔高度与无线电高度差，用于校正高度
    off_qnh = alt_qnh[int(flare_point/8)] - ralt_mean[int(flare_point/2)]  #海拔高度与无线电高度差，用于校正高度
    dic['ALT_STD'] = list(np.array(alt_std) - off_std)  #校正后的高度(适用于高度大于50ft时)
    dic['ALT_QNH'] = list(np.array(alt_qnh) - off_qnh)  #校正后的高度(适用于高度大于50ft时)
    
    dic['VRTG'] = getDataList(df, 'VRTG', 8)  #垂直载荷
    #dic['LAT_ACC'] = getDataList(df, 'LATG_C', 4) #横向加速度
    #dic['LON_ACC'] = getDataList(df, 'LONG_C', 4) #纵向加速度
    dic['LON'] = list(df['LONPC'])  #经度
    dic['LAT'] = list(df['LATPC'])  #维度
    dic['GW'] = list(df['GW_C'])  #毛重
    
    dic['IAS'] = list(df['IAS']) #计算空速
    dic['VAPP'] = list(df['VAPP'])  #速度参考
    dic['GS'] = list(df['GS'])   #地速
    dic['IVV'] = list(df['IVV_CA'])   #垂直速度
    
    head_mag = toFloat(list(df['HEAD_MAG']))
    win_dir = toFloat(list(df['WIN_DIR']))  #风向
    win_spd = list(df['WIN_SPD'])  #风速
    cross_deg = (np.array(win_dir).astype('float64') - np.array(head_mag).astype('float64')) * np.pi / 180
    dic['WIN_DIR'], dic['WIN_SPD'] = win_dir, win_spd
    dic['WIN_CRS'] = list(np.array(win_spd) * np.sin(cross_deg))  #侧向风
    dic['WIN_ALG'] = list(np.array(win_spd) * np.cos(cross_deg))  #纵向风
    
    dic['PITCH'] = getDataList(df, 'PITCH', 4)  #俯仰
    dic['PITCH_CPT'] = toFloat(getDataList1(df, 'PITCH_CPT', 8, flag))
    dic['PITCH_FO'] = toFloat(getDataList1(df, 'PITCH_FO', 8, flag))    
    dic['PITCH_CMD'] = list(np.array(dic['PITCH_CPT']) + np.array(dic['PITCH_FO']))
    
    dic['ROLL'] = getDataList(df, 'ROLL', 2)  #横滚
    dic['ROLL_CPT'] = toFloat(getDataList1(df, 'ROLL_CPT', 8, flag))
    dic['ROLL_FO'] = toFloat(getDataList1(df, 'ROLL_FO', 8, flag))
    dic['ROLL_CMD'] = list(np.array(dic['ROLL_CPT']) + np.array(dic['ROLL_FO']))
    
    #根据降落阶段主操作员将另外没有操作的人员相关词操作参数设置为Nan
    if majorOperator(dic['PITCH_CPT'][flare_point:point], dic['PITCH_FO'][flare_point:point]) == 'First':
        dic['PITCH_FO'] = list(np.full(len(dic['PITCH_FO']), np.nan)) #若为机长操作，则副机长设为Nan
        dic['ROLL_FO'] = list(np.full(len(dic['ROLL_FO']), np.nan)) #若为机长操作，则副机长设为Nan
    else:
        dic['PITCH_CPT'] = list(np.full(len(dic['PITCH_CPT']), np.nan)) #若为副机长操作，机长设为Nan
        dic['ROLL_CPT'] = list(np.full(len(dic['ROLL_CPT']), np.nan)) #若为副机长操作，机长设为Nan
        
    dic['N11'] = list(df['N11'])  #发动机转速比
    dic['N12'] = list(df['N12'])
    dic['TLA1'] = list(df['TLA1'])  #油门控制杆角度
    dic['TLA2'] = list(df['TLA2'])
    
    dic['FLAP_PL'] = fillNan(list(df['FLAP_POS_LH']))
    dic['FLAP_PR'] = fillNan(list(df['FLAP_POS_RH']))
    dic['FLAP_C'] = list(df['FLAPC'])  #襟翼位置
    
    dic['RUDD'] = getDataList1(df, 'RUDD', 2, flag)  #方向舵位置
    
    #dic['GLIDE_DEV'] = list(df['GLIDE_DEV1_C']) #下滑道偏差
    #dic['LOC_DEV'] = list(df['LOC_DEV1_C']) #航向台偏差
    dic['DME1'] = fillNan(list(df['DME1DSTD']))  #航向偏离角
    dic['DME2'] = fillNan(list(df['DME2DSTD'])) 
    
    #dic['STAB'] = getDataList(df, 'STAB', 2) #安定面位置
    
    return dic
    #outPath = 'D:/A320/Data50s/'+file.replace(".csv","_50s.csv")
    #outDf = pd.DataFrame(dic)
    #outDf.to_csv(outPath,index=False)
    
    
#=======================================================多个数据源合并====================================================    

source = ['/mnt/sdb/Dataset/QarBigData/Chengdu/All/',"/mnt/sdb/Dataset/QarBigData/HubeiA320/"]
pathList = []
for sourcepath in source:
    pathList.extend(list_all_files(sourcepath,'.csv'))
print(len(pathList))

#=======================================================提取50s信息========================================================

info,filelist = [],[]
for path in tqdm(pathList):
    dic = computeInfo(path)
    pathsplit = path.split('/')
    file = pathsplit[len(pathsplit)-1]
    filelist.append(file)
    info.append(dic)

outPath = 'Info100s.csv'
outDf =pd.DataFrame({"FILE":filelist,"INFO":info})
outDf.to_csv(outPath,index=False)     
```

```python
import os

def list_all_files(rootdir, ext=None, excpt=None):
    import os
    _files = []
    list = os.listdir(rootdir) #列出文件夹下所有的目录与文件
    for i in range(0,len(list)):
        path = os.path.join(rootdir,list[i])
        if os.path.isdir(path) and (excpt == None or os.path.basename(path) not in excpt):
            _files.extend(list_all_files(path, ext))
        if os.path.isfile(path) and (ext == None or ext in path):
            _files.append(path)
    return _files

def filterNan(dataList):
    import numpy as np
    newList = []
    for i in range(len(dataList)):
        if not np.isnan(float(dataList[i])):
            newList.append(float(dataList[i]))
    return newList

def analysis(path, statDir):
    import pandas as pd
    import numpy as np
    df = pd.read_csv(path)
    mean,std = [], []
    rank10, rank90, rank05, rank95, rank025, rank975 = [], [], [], [], [], []
    for col in df:
        dataList = filterNan(list(df[col]))
        if path.find('_CPT') != -1 or path.find('_FO') != -1:
            dataList = filterNan(dataList) #过滤掉列表中的NaN值
        size = len(dataList)
        mean.append(np.mean(dataList)) #均值
        std.append(np.std(dataList)) #方差
        
        dataList.sort() #对数据排序
        rank10.append(dataList[int(size*0.1)]) #80%区间下界
        rank90.append(dataList[int(size*0.9)]) #80%区间上界
        rank05.append(dataList[int(size*0.05)]) #90%区间下界
        rank95.append(dataList[int(size*0.95)]) #90%区间上界
        rank025.append(dataList[int(size*0.025)]) #95%区间下界
        rank975.append(dataList[int(size*0.975)]) #95%区间上界
    outDf = pd.DataFrame({'MEAN': mean, 'STD': std, 'R10': rank10, 'R90':rank90, 
                         'R05':rank05, 'R95':rank95, 'R025':rank025, 'R975':rank975})
    outPath = statDir + '/' + os.path.basename(path).replace('.csv', '_stat.csv')
    outDf.to_csv(outPath, index=False)

baseDir = "/mnt/sdb/Dataset/LiCX/A320/params"
statDir = "/mnt/sdb/Dataset/LiCX/A320/statistics"
pathList = list_all_files(baseDir, '.csv')
counter = 0
print(len(pathList), '文件被处理中:')
for i in range(0, len(pathList)):
    counter += 1
    path = pathList[i]
    print(os.path.basename(path), end=' ')
    analysis(path, statDir)
print('\n', counter, '文件生成成功!')
print('成功!')
```

```python
import numpy as np

# 读取多个CSV文件，分别保存不同的参数数据
ldgl_df = pd.read_csv("/mnt/sdb/Dataset/LiCX/A320/params/LDGL.csv")
ldgnos_df = pd.read_csv("/mnt/sdb/Dataset/LiCX/A320/params/LDGNOS.csv")
ldgr_df = pd.read_csv("/mnt/sdb/Dataset/LiCX/A320/params/LDGR.csv")
vrtg_df = pd.read_csv("/mnt/sdb/Dataset/LiCX/A320/params/VRTG.csv")
pitch_df = pd.read_csv("/mnt/sdb/Dataset/LiCX/A320/params/PITCH.csv")
ralt_mean_df = pd.read_csv("/mnt/sdb/Dataset/LiCX/A320/params/RALT_MEAN.csv")

# 初始化计数器和用于记录各种情况的空列表
count = 0  # 统计异常最大VRTG出现在着陆前25秒的次数
count_list = []  # 记录最大VRTG出现在接近着陆点时的所有数据
norm_list_hl = []  # 正常着陆，最大VRTG出现在50ft到完全接地之间
abnorm_list_hl = []  # 异常着陆，最大VRTG出现在50ft之前或未接地
norm_list_nl = []  # 无效着陆，最大VRTG出现在50ft到完全接地之间
abnorm_list_nl = []  # 无效着陆，最大VRTG出现在50ft之前或未接地
time_list = []  # 记录最大VRTG或大于1.5的地方的时间

# 初始化用于记录其他分析数据的空列表
ground_list = []
fp_list, td_list, ctd_list, ft_list = [], [], [], []

# 遍历VRTG数据集的每一行
for i in range(len(vrtg_df)):
    # 获取每个文件的当前行数据
    ldgl = ldgl_df.values[i].tolist()
    ldgnos = ldgnos_df.values[i].tolist()
    ldgr = ldgr_df.values[i].tolist()
    vrtg = vrtg_df.values[i].tolist()
    ralt_mean = ralt_mean_df.values[i].tolist()
    
    # 计算离50ft高度最近的点的位置（即起飞点）
    diffList = list(np.abs(np.array(ralt_mean)-50))  
    flare_point = 2 * diffList.index(min(diffList))  # Flare point 是接地时的浮空点位置
    
    # 获取着陆点（精度为0.25秒）
    point = 2 * min(ldgl.index(0), ldgr.index(0), ldgnos.index(0)) 
    
    # 查找着陆点在数据中的位置
    ground = len(ldgnos) - 1
    while ldgnos[ground] == 1:
        ground -= 1
    while ldgnos[ground] == 0:
        ground -= 1
    
    ground_list.append(ground)
    
    # 计算flare_time，表示起飞与接地之间的时间差
    flare_time = (point - flare_point) / 8
    
    # 记录flare point, touchdown point, contact time, flare time
    fp_list.append(flare_point)
    td_list.append(point)
    ctd_list.append(ground)
    ft_list.append(flare_time)
    
    # 查找VRTG的最大值
    m = max(vrtg)
    if m < 1.5:
        # 如果最大VRTG小于1.5，判断它的位置
        if m >= 1.4:
            # 倒序遍历，寻找VRTG大于1.4的位置
            for j in range(len(vrtg)-1, -1, -1): 
                if vrtg[j] >= 1.4:
                    time_list.append(j)
                    # 如果最大VRTG出现在50ft到接地之间
                    if j <= ground and j > flare_point: 
                        norm_list_nl.append(i)
                    else:
                        count += 1  # 统计异常情况
                        print("NL")  # 输出标记为异常着陆
                        abnorm_list_nl.append(i)
                        count_list.append(vrtg)  # 记录异常VRTG
                    break
        else:
            norm_list_nl.append(i)
            time_list.append(ground)
    else:
        # 如果最大VRTG大于等于1.5，继续查找其位置
        for j in range(len(vrtg)-1, -1, -1):
            if vrtg[j] >= 1.5:
                time_list.append(j)
                # 如果最大VRTG出现在50ft到接地之间
                if j <= ground and j > flare_point: 
                    norm_list_hl.append(i)
                else:
                    count += 1  # 统计异常情况
                    print("HL")  # 输出标记为异常着陆
                    abnorm_list_hl.append(i)
                    count_list.append(vrtg)  # 记录异常VRTG
                break
```

```python
import warnings
warnings.filterwarnings('ignore')
import numpy as np
import pandas as pd
from numpy import array as array
from numpy import nan as nan
import os
import matplotlib.pyplot as plt
from torch import nn
from torch.nn.utils import weight_norm
from sklearn.model_selection import train_test_split
import torch
from torch.autograd import Variable
import torch.optim as optim
import torch.nn.functional as F
import sys
import argparse
import imblearn
from imblearn.over_sampling import RandomOverSampler
from collections import Counter
import seaborn as sns
from sklearn import preprocessing
from tqdm import tqdm

class Chomp1d(nn.Module):
    def __init__(self, chomp_size):
        super(Chomp1d, self).__init__()
        self.chomp_size = chomp_size

    def forward(self, x):
        return x[:, :, :-self.chomp_size].contiguous()


class TemporalBlock(nn.Module):
    def __init__(self, n_inputs, n_outputs, kernel_size, stride, dilation, padding, dropout=0.2):
        super(TemporalBlock, self).__init__()
        self.conv1 = weight_norm(nn.Conv1d(n_inputs, n_outputs, kernel_size,
                                           stride=stride, padding=padding, dilation=dilation))
        self.chomp1 = Chomp1d(padding)
        self.relu1 = nn.ReLU()
        self.dropout1 = nn.Dropout(dropout)

        self.conv2 = weight_norm(nn.Conv1d(n_outputs, n_outputs, kernel_size,
                                           stride=stride, padding=padding, dilation=dilation))
        self.chomp2 = Chomp1d(padding)
        self.relu2 = nn.ReLU()
        self.dropout2 = nn.Dropout(dropout)

        self.net = nn.Sequential(self.conv1, self.chomp1, self.relu1, self.conv2, self.chomp2, self.dropout1)
                                 #, self.relu2, self.dropout2)
        self.downsample = nn.Conv1d(n_inputs, n_outputs, 1) if n_inputs != n_outputs else None
        self.relu = nn.ReLU()
        self.init_weights()

    def init_weights(self):
        self.conv1.weight.data.normal_(0, 0.01)
        self.conv2.weight.data.normal_(0, 0.01)
        if self.downsample is not None:
            self.downsample.weight.data.normal_(0, 0.01)

    def forward(self, x):
        out = self.net(x)
        res = x if self.downsample is None else self.downsample(x)
        return self.relu(out + res)


class TemporalConvNet(nn.Module):
    def __init__(self, num_inputs, num_channels, kernel_size=2, dropout=0.2):
        super(TemporalConvNet, self).__init__()
        layers = []
        #print(num_inputs)
        num_levels = len(num_channels)
        for i in range(num_levels):
            dilation_size = 2 ** i
            #print(dilation_size)
            in_channels = num_inputs if i == 0 else num_channels[i-1]
            out_channels = num_channels[i]
            #print(i,dilation_size,in_channels,out_channels)
            layers += [TemporalBlock(in_channels, out_channels, kernel_size, stride=1, dilation=dilation_size,
                                     padding=(kernel_size-1) * dilation_size, dropout=dropout)]

        self.network = nn.Sequential(*layers)

    def forward(self, x):
        return self.network(x)
```
